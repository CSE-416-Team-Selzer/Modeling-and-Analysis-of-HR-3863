```python
import geopandas as gpd
import pandas as pd
import jsonpickle as jp
import json
pd.set_option('display.max_columns', None)
```

Importing Data

```python
bound = gpd.read_file('data/Utah/boundary_data/ut_2020.shp')
usVote = pd.read_csv('data/2020-HOUSE-precinct-general.csv')
```

Boundary Data Preprocessing

```python
bound = bound[bound.columns.drop(list(bound.filter(regex='G20')))]
#bound.drop(["CDE_COUNTY","G20PREDBID","G20PRERTRU","G20PRELJOR","G20USSDKEL","G20USSRMCS"], axis = 1, inplace=True)
bound['resultspre'].fillna('na', inplace = True)
bound['resultspre'] = bound['resultspre'].str.lower()
```

Voting Data preprocessing

```python
usVote.drop(["office",'party_detailed','mode','county_name','county_fips','jurisdiction_name','jurisdiction_fips','candidate','district','magnitude','dataverse','year','stage','state','special','writein','state_fips','state_cen','state_ic','date','readme_check'], axis = 1, inplace=True)
vote = usVote[usVote['state_po'] == 'UT'].copy()
vote.drop(["state_po"], axis = 1, inplace=True)
vote['precinct'] = vote['precinct'].str.lower()
display(vote)
```

Combining Data Functions

```python
def total_votes(precinct, party, source):
    if (precinct == 'na'):
        return 0
    #precintdf = source[(source['precinct'].str.contains(precinct)) & (source['party'] == party)]
    precinctdf = get_precinct(precinct, source)
    precinctdf = precinctdf[precinctdf['party_simplified'] == party]
    return precinctdf['votes'].sum()

def get_precinct(precinct, source):
    precinctdf = source[source['precinct'].str.contains(precinct)]
    return precinctdf
```

Combining Data

```python
state = bound
#display(state)
#display(state[state['resultspre'].str.contains('beaver')])
display(state[state['resultspre'] == 'na'])
display(vote[vote['precinct'].str.contains('bo03')])
display(state.plot())
```

```python
state['demVotes'] = state['resultspre'].apply(lambda precinct: total_votes(precinct, 'DEMOCRAT', vote))
state['repVotes'] = state['resultspre'].apply(lambda precinct: total_votes(precinct, 'REPUBLICAN', vote))
state['totalVotes'] = state['demVotes'] + state['repVotes']
display(state)
```

```python
display(state[state['resultspre'].str.contains('unincorporated')])
display(vote[vote['precinct'].str.contains('unincorporated')])
display(state['totalVotes'].value_counts())
display(state[state['totalVotes'] == 0])
```

Exporting State Data

```python
state.to_file('data/Utah/utahGeo.geojson', driver='GeoJSON', index = True)
noGeo = pd.DataFrame(state.drop(columns='geometry'))
noGeo.to_json('data/Utah/utah.json', orient = 'index', compression = 'infer', index = True, indent = 4)
```
