```python
import geopandas as gpd
import pandas as pd
import json
import math
```

```python
with open('data/Arizona/ensamble/smd_ensemble_Arizona.json') as json_file:
    ensamble_ids = json.load(json_file)
precincts = gpd.read_file('data/Arizona/arizonaGeo.geojson')
display(type(ensamble_ids))
display(precincts)
```

```python
display(type(ensamble_ids['0']))
display(ensamble_ids['0'])
ensamble = dict()
```

```python
def polsby_popper(area, perimeter):
    return (4 * math.pi * area) / (perimeter ** 2)
```

```python
def map_to_district(precinct_id, plan_id):
    plan = ensamble_ids[plan_id]
    #display(precinct_id)
    #display(plan)
    for district_id in plan:
        #display(district_id)
        if(precinct_id in plan[district_id]):
            return int(district_id)
    return -1

for plan_id in ensamble_ids:
    display(plan_id)
    district_plan = precincts.copy(deep = True)
    district_plan['district'] = district_plan['index'].apply(lambda id: map_to_district(id, plan_id))
    #display(district_plan)
    #display(district_plan.plot())
    district_plan.drop(["index", "precinct"], axis = 1, inplace=True)
    district_plan = district_plan.dissolve(by = 'district', aggfunc = 'sum')
    #display(district_plan)
    district_plan['id'] = district_plan.index
    #district_plan['color'] = '#C62828'
    #district_plan.loc[6, 'color'] = '#283593'
    #display(district_plan.plot(color = district_plan['color']))
    #display(district_plan.iloc[1].geometry.plot())
    ensamble[int(plan_id)] = district_plan
    if(plan_id == '12'):
        break
```

```python
display(len(ensamble))
display(ensamble[0].plot())
display(ensamble[0])
display(ensamble[5].plot())
display(ensamble[5])
display(ensamble[10].plot())
display(ensamble[10])
```
